正規乱数を生成するプログラムを作れ。  
また、そのプログラムを用いて $𝜇 = 0$, $𝜎 = 1$ の正規分布に従う正規分布を1000個生成し、その正規確率をヒストグラムで表示せよ。  
階級数は30とする。

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np

# 正規乱数を生成(𝜇 = 0, 𝜎 = 1)
x = np.random.normal(0, 1, 1000)
x = np.random.normal(0, 1, 1000, )

# 表示
plt.hist(x, 30)
plt.xlabel('x')
plt.ylabel('freq')
plt.show()
```
